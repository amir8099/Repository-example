## Question 1-Extracting Table form Wikipedia page

```python
!pip install lxml
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source)
print(soup.title)
from IPython.display import display_html
table1 = str(soup.table)
display_html(table1,raw=True)
```

## Coverting teh table to DataFrame

```python
dfs = pd.read_html(table1)
df=dfs[0]
df.head()
```

## Wrangling the data

```python
# Dropping the rows where Borough is 'Not assigned'
df1 = df[df.Borough != 'Not assigned']

# Combining the neighbourhoods with same postal code
df_group=df1.groupby(['Postal Code','Borough'], sort=False).agg(', '.join)
df_group.reset_index(inplace=True)

#Replacing the name of the neighbourhoods which are 'Not assigned' with names of Borough
df_group.loc[df_group['Neighborhood'] == 'Not assigned', 'Neighborhood'] = df_group['Borough']
df_group.head()
```

```python
#Shape of dataframe
df_group.shape
```

## Question 2
## Importing the csv file-latitudes and longitudes

```python
lat_lon = pd.read_csv('https://cocl.us/Geospatial_data')
lat_lon.head()
```

```python
df_geo = pd.merge(df_group,lat_lon,on='Postal Code')
df_geo.head()
```

## Clustering boroughs that contain the word Toronto

```python
df_toronto = df_geo[df_geo['Borough'].str.contains('Toronto',regex=False)]
df_toronto
```

```python
## Cluster Neighborohoods in Toronto based on their Latitute and Longitude Information
```

```python
# set number of clusters
kclusters = 5

toronto_grouped_clustering = df_toronto.drop(['Postal Code','Borough', 'Neighborhood'], 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
df_toronto
```

## Map of Clusters

```python
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighborhood, cluster in zip(df_toronto['Latitude'], df_toronto['Longitude'], df_toronto['Neighborhood'], df_toronto['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
